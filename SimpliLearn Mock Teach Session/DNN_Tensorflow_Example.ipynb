## We will be working with the MNIST dataset.

The **MNIST dataset (Modified National Institute of Standards and Technology)** is a classic benchmark dataset in machine learning, especially for image classification tasks involving handwritten digits. It was created by Yann LeCun, Corinna Cortes, and Chris Burges by remixing samples from the original NIST datasets.

**Dataset Contents:**

* 70,000 total grayscale images
* 60,000 training images
* 10,000 test images

**Image properties:**

* Size: 28 × 28 pixels
* Format: Grayscale, with pixel intensities from 0 (white) to 255 (black)
* Each image flattened into 784 features when used in vector form

**Labels:**

* Digits 0 through 9 (10 classes)

## Import TensorFlow and load the MNIST dataset.

The data will be preprocessed by flattening the images into 1D arrays.

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist
import warnings
warnings.filterwarnings('ignore')

# Load and preprocess the MNIST data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0        # Normalize pixel values to be between 0 and 1

# Get shapes of input matrices and output vectors
print('x_train: ',x_train.shape,'\nx_test: ',x_test.shape, '\ny_train: ',y_train.shape,'\ny_test: ',y_test.shape)
```

## Use the `tf.keras.Sequential` API to stack the layers of the DNN.

* **Flatten**: Converts the 28x28 image input into a 784-element 1D vector.
* **Dense**: A fully connected layer of neurons.

The final Dense layer has 10 outputs (for 10 classes) and uses the softmax activation to provide probability scores for each class.

```python
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')  # outputs probabilities for digits 0-9
])

model.summary()   # get summary information about the model we just built
```

## Configure the model's learning process using an optimizer, a loss function, and metrics

```python
model.compile(optimizer='adam',       # Adam (Adaptive Moment Estimation) is one of the most popular optimizers.
                                      # It adapts the learning rate for each parameter, making training faster and more stable.
                                      # Great default choice for most deep learning tasks.

              loss='sparse_categorical_crossentropy',  # `sparse_categorical_crossentropy` is used for multi-class classification problems.
              metrics=['accuracy'])                    # `accuracy` tells you what percentage of predictions were correct.
```


## Fit the training data to the model. The epochs parameter defines how many times the model will see the entire dataset.

```python
model.fit(x_train, y_train, epochs=5)
```

## Output the model predictions for the test dataset

```python
model.predict(x_test)
```

## Check the model's performance on the test data to see how well it generalizes to unseen examples.

```python
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)   # `model.evaluate()` computes the loss and metrics of
                                                                  # a trained model on a given dataset. It is typically used on
                                                                  # validation or test data to measure how well the model generalizes.

print(f'\nTest accuracy: {test_acc}')
```
